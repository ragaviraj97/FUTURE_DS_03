```python
import pandas as pd
df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")
visitors = 50000
leads = int(visitors * 0.2)
customers = df['Customer Name'].nunique()

funnel = pd.DataFrame({
    "Stage": ["Visitors", "Leads", "Customers"],
    "Count": [visitors, leads, customers]
})
funnel
```

```python
# conversion rates
traffic_to_leads = (leads/visitors)* 100
leads_to_customers =(customers/leads)*100
overall_conversion = (customers/visitors)*100
print("Traffic + Leads:",round(traffic_to_leads,2),"%")
print("Leads + Customers:",round(leads_to_customers,2),"%")
print("Overall_Conversion",round(overall_conversion,2),"%")
```

```python
# Drop from Visitors
drop_visitors = ((visitors - leads) / visitors) * 100 if visitors != 0 else 0

# Drop from Leads
if leads == 0:
    drop_leads = 0
elif customers > leads:
    print("!!! Data Issue: Customers cannot be greater than Leads")
    drop_leads = 0
else:
    drop_leads = ((leads - customers) / leads) * 100


print("Drop from Visitors:", round(drop_visitors, 2), "%")
print("Drop from Leads:", round(drop_leads, 2), "%")
```

```python
# Funnel Table
funnel = pd.DataFrame({
    "Stage": ["Visitors", "Leads", "Customers"],
    "count": [visitors,leads,customers]
})
funnel
```

```python
import matplotlib.pyplot as plt
plt.bar(funnel["Stage"],funnel["count"],color="skyblue")
plt.title("Funnel Analysis")
plt.xlabel("Stage")
plt.ylabel("Users")
plt.show()
```

```python
region_analysis= df.groupby("Region")["Customer Name"].nunique().sort_values(ascending=False)
print(region_analysis)
```

```python
region_analysis.plot(kind='pie', autopct='%1.1f%%')

plt.title("Customer Distribution by Region")
plt.ylabel("")
plt.show()
```

```python
# Category performance
category_analysis= df.groupby("Category")["Sales"].sum()
print(category_analysis)
```

```python
category_analysis.sort_values().plot(kind='barh')

plt.title("Sales by Category")
plt.xlabel("Total Sales")
plt.ylabel("Category")
plt.show()
```

```python
 # Time-Based Analysis

df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = (
    df.groupby(df['Order Date'].dt.to_period('M'))['Sales']
    .sum()
    .sort_index()
)

print(monthly_sales)
monthly_sales.to_csv("monthly_sales.csv")

from google.colab import files
files.download("monthly_sales.csv")
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
